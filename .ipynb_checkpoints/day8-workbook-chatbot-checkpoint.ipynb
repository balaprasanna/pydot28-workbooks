````python
import seaaborn as sns
df = sns.load_dataset("tips")
df.head()
````

````python
fig, ax = plt.subplots(1,1)
# sns.scatterplot(x="total_bill", y="tip", data=df, hue="sex", ax=ax[2])

male_x = df.query("sex=='Male'")["total_bill"].values
male_y = df.query("sex=='Male'")["tip"].values

female_x = df.query("sex=='Female'")["total_bill"].values
female_y = df.query("sex=='Female'")["tip"].values

ax.set_xlabel("total_bill")
ax.set_ylabel("tip")

plt.scatter(male_x,male_y, marker="*", label="Male",color="orange")
plt.scatter(female_x,female_y, marker="8", label="Female",color="green")
# ax.legend()
````

````python
fig, ax = plt.subplots(1,2, figsize=(10,4))

fig.suptitle('Main title', x=0.5,y=1.03)

male_x = df.query("sex=='Male'")["total_bill"].values
male_y = df.query("sex=='Male'")["tip"].values

female_x = df.query("sex=='Female'")["total_bill"].values
female_y = df.query("sex=='Female'")["tip"].values

ax[0].set_xlabel("total_bill")
ax[0].set_ylabel("tip")

ax[1].set_xlabel("total_bill")
ax[1].set_ylabel("tip")

ax[0].scatter(male_x,male_y, marker="*", label="Male",color="orange" )
ax[1].scatter(female_x,female_y, marker="8", label="Female",color="green")
# ax.legend()
plt.tight_layout()
````

````python
res = df.groupby("day")['tip'].mean()
x = res.index
y = res.values
````

````python
# library
import matplotlib.pyplot as plt
from matplotlib import colors as c
 
# create data
names=x
size=y

colors = [c[0] for c in list(c.cnames.items())[:len(y)]]
print(colors)
 
# Create a circle for the center of the plot
my_circle=plt.Circle( (0,0), 0.5, color='white')

# Custom colors --> colors will cycle
plt.pie(size, labels=names, colors=colors)
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
````
